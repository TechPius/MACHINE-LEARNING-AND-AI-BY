Optimizing K-value of K-NN model

```python
# Run a for-loop of many possible k-values and see the lowest k-value with highest accuracy

scores=[]
for k in range(1,25):
    knn_test=KNeighborsClassifier(n_neighbors=k)
    knn_test.fit(Kx_train,Ky_train)
    Ky_pred=knn_test.predict(Kx_test)
    print('Accuracy for k=',k,'is:',round(accuracy_score(Ky_pred,Ky_test),2))
    scores.append(round(accuracy_score(Ky_pred,Ky_test),3))
```

Visualize optimal K-value accuracy

```python
plt.plot(range(1,25),scores)
plt.xlabel('Value K for KNN')
plt.ylabel('Testing Accuracy')
plt.vlines(x=14, ymin=min(scores), ymax=max(scores), color='red')
plt.show()
```

```python
# Data preprocessing already done in KNN

# recall features and the target variable (iris_X and iris_y)

# Standardize the features using Standard scaler
from sklearn.preprocessing import StandardScaler

col = iris_X.columns  #save the original column headers
scaler = StandardScaler()  #initialize the scaler
ScaledX = scaler.fit_transform(iris_X)    #transform features to normalized form
Scal_Iris_X = pd.DataFrame(ScaledX, columns=col) #recreate a dataframe of scaled features
```

```python
Scal_Iris_X.sample(15)
```

```python
# Split the data into training and test sets for use in SVM classification
SX_train, SX_test, Sy_train, Sy_test = train_test_split(Scal_Iris_X, iris_y, test_size=0.2, random_state=42)
```

```python
# Import SVM classifier
from sklearn.svm import SVC

# Initialize the SVM model
svm_model = SVC(kernel='linear', random_state=42)

# Train the model using the training data
svm_model.fit(SX_train, Sy_train)

# Make predictions on the test data
Sy_pred = svm_model.predict(SX_test)

# Evaluate the model
accuracy_svm = accuracy_score(Sy_test, Sy_pred)
report_svm = classification_report(Sy_test, Sy_pred)
conf_matrix_svm = confusion_matrix(Sy_test, Sy_pred)

print(f"SVM Accuracy: {accuracy_svm}")
print("SVM Classification Report:")
print(report_svm)
```

To improve the accuracy of our prediction with SVC we can gridsearchCV

Practical Explanation of hyperparameters in SVM

- C: Imagine you are building a fence to separate different animals in a farm. A low C value means you allow some animals to be misclassified (the fence is more flexible), while a high C value means you try to classify every animal correctly, which may make the fence very complex and overfitted to the specific arrangement of animals in your training set.

- gamma: If gamma is low, each animal affects a large area around it, leading to a smoother decision boundary. If gamma is high, each animal affects only a small area around it, leading to a more complex decision boundary that might fit the noise in your data.

- kernel: Choosing a kernel is like choosing the shape of the fence. A linear kernel means a straight fence, while an RBF kernel means a flexible, wavy fence that can bend to better separate the animals.

```python
# Import necessary libraries
from sklearn.model_selection import GridSearchCV

# Define the parameter grid for GridSearchCV
parameter_grid = {
    'C': [0.1, 1, 10, 100],
    'gamma': [1, 0.1, 0.01, 0.001],
    'kernel': ['linear', 'rbf', 'poly', 'sigmoid']
}

# Initialize GridSearchCV with the SVM model
grid_search = GridSearchCV(estimator=SVC(),  param_grid=parameter_grid, refit=True, cv=5)

# Train the model using GridSearchCV
grid_search.fit(SX_train, Sy_train)

# Get the best model from GridSearchCV
best_svm_model = grid_search.best_estimator_
```

```python
best_svm_model
```

```python
# Make predictions on the test data using the best model
besty_pred = best_svm_model.predict(SX_test)

# Evaluate the best model
accuracy_svm = accuracy_score(Sy_test, besty_pred)
report_svm = classification_report(Sy_test, besty_pred)
conf_matrix_svm = confusion_matrix(Sy_test, besty_pred)

print(f"SVM Accuracy: {accuracy_svm}")
print("SVM Classification Report:")
print(report_svm)

# Plot the confusion matrix
plt.figure(figsize=(8, 6))
sns.heatmap(conf_matrix_svm, annot=True,
            xticklabels=le.classes_,
            yticklabels=le.classes_)
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('SVM Confusion Matrix')
plt.show()
```
